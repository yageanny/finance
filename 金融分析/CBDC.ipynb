```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

# 对于每个项目的国家列表，将列表中的国家视为连接的节点添加到图中
for country_list in project_countries:
    # 如果项目中有多个国家，则在它们之间创建边
    if len(country_list) > 1:
        # 添加所有可能的边
        G.add_edges_from([(country_list[i], country_list[j]) for i in range(len(country_list)) for j in range(i + 1, len(country_list))])

# 现在你可以进行网络计算，比如计算度、中心性等
# 例如，计算每个节点的度
degrees = dict(G.degree())

# 创建一个颜色列表，颜色由节点的度决定，度数越高，颜色越深
colors = [degrees[node] for node in G.nodes()]

# 创建一个节点大小列表，节点大小由节点的度决定
sizes = [degrees[node] * 50 for node in G.nodes()]

# 使用spring_layout布局使图更紧凑， k值越小，图越紧凑
pos = nx.spring_layout(G, k=0.60)

plt.figure(figsize=(10, 5))  # 设置图形大小

# 绘制网络图，设置标题和颜色图例，选择YlOrBr颜色图
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=sizes, edge_color='k', linewidths=1, font_size=10, cmap=plt.cm.YlOrBr)
plt.title('CBDC Network Graph', fontsize=10)

# 创建colorbar并设置其大小
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.YlOrBr), label='Degree', aspect=30, shrink=0.5)
cbar.set_label('Degree', size=10)  # 设置colorbar的标签大小

plt.show()
```

```python


# 计算并打印网络特征
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)
transitivity = nx.transitivity(G)
average_degree = sum(degrees.values()) / num_nodes
clustering_coefficient = nx.average_clustering(G)

print("Size (No. of Nodes):", num_nodes)
print("Size (No. of Edges):", num_edges)
print("Density:", density)
print("Transitivity:", transitivity)
print("Average Degree:", average_degree)
print("Clustering Coefficient:", clustering_coefficient)

# 计算中心化指标
out_degree_centralization = nx.degree_centrality(G)
in_degree_centralization = nx.degree_centrality(G)  # 因为无向图所以入度和出度中心化是相同的

print("Out-Degree Centralization:", max(out_degree_centralization.values()))
print("In-Degree Centralization:", max(in_degree_centralization.values()))

# Reciprocity is often used for directed graphs, but for undirected graphs it's always 1.0
reciprocity = 1.0 if nx.is_directed(G) else "N/A"

print("Reciprocity:", reciprocity)
```

```python
# 去掉空字符
def delete_r_n(line):
    return line.replace('\r', '').replace('\n', '').strip()

# 加载停用词
def get_stop_words(stop_words_dir):
    stop_words = []

    with open(stop_words_dir, 'r', encoding='utf-8') as f_reader:
        for line in f_reader:
            line = delete_r_n(line)
            stop_words.append(line)

    stop_words = set(stop_words)
    return stop_words

# 加载停用词,stopwords为停用词列表
stopwords = get_stop_words(r'D:/作业/stop1.txt')

```
